#### What's the mean salary by department?

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
employee.groupby(["Department", "Title"]).agg({"Salary":["mean","count"]}).reset_index()
```

```python
employee.pivot_table(values = "Salary", index=["Department", "Title"], aggfunc=["mean","count"]).reset_index()
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
employee.groupby(["Department", "Title"]).agg({"Salary":[np.median, min, max, np.std]}).reset_index()
```

```python
salary_stats = employee.pivot_table(values = "Salary", index = ["Department", "Title"], aggfunc = [np.median, min, max, np.std]).reset_index()
salary_stats
```

Salaries are pretty much the same accross employees with the same title, VP is the job that earns the most overall and analyst the one that earns the lowest (sad).

Missing values in std is due to only having 1 employee.

#### Based on your comments, fill the missing values with an appropriate value.

```python
salary_stats[('std', 'Salary')] = salary_stats[('std', 'Salary')].fillna(0)
salary_stats
```

### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
# In this case, the Sales Department will dissappear due to both employees earning the same, we could avoid including it in the function or not taking it into account as the mean is absurd to calculate.

def not_at_the_top(df):
    '''Removes the top earner(s) for each department and produces a pivot table
    with the mean salary per department.
    Parameters: A DataFrame
    Returns: a pivot table with the mean salary per department'''
    
    df1 = df[df["Department"] == "HR"].copy()
    df2 = df[df["Department"] == "IT"].copy()
    df3 = df[df["Department"] == "Sales"].copy() # Optional
    
    df1 = df1[df1["Salary"] != df1["Salary"].max()]
    df2 = df2[df2["Salary"] != df2["Salary"].max()]
    df3 = df3[df3["Salary"] != df3["Salary"].max()] # Optional
    
    df4 = pd.concat([df1,df2,df3], axis = 0)
    
    # df4pt = df4.groupby(["Department"]).agg({"Salary":"mean"})
    df4pt = df4.pivot_table(values = "Salary", index = "Department").reset_index()
                 
    return df4pt
    
not_at_the_top(employee)
```
